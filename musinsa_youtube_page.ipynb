```python
import requests
# import json
import pandas as pd
from time import sleep

# âœ… 1. ê¸°ë³¸ URL
url = "https://api.musinsa.com/api2/dp/v1/plp/goods"

# âœ… 2. í—¤ë” (ë¸Œë¼ìš°ì € ìš”ì²­ê³¼ ë™ì¼í•˜ê²Œ ì„¤ì •)
headers = {
    "authority": "api.musinsa.com",
    "method": "GET",
    "scheme": "https",
    "accept": "application/json, text/plain, */*",
    "accept-encoding": "gzip, deflate, br, zstd",
    "accept-language": "ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7",
    "origin": "https://www.musinsa.com",
    "referer": "https://www.musinsa.com/",
    "sec-fetch-dest": "empty",
    "sec-fetch-mode": "cors",
    "sec-fetch-site": "same-site",
    "user-agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/118.0.0.0 Safari/537.36"
    ),
}

# âœ… 3. í¬ë¡¤ë§ ê²°ê³¼ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
all_items = []

# âœ… 4. í˜ì´ì§€ 1~8 ë°˜ë³µ
for page in range(1, 9):
    params = {
        "gf": "A",
        "keyword": "ë¬´ì‹ ì‚¬Ã—ìœ íŠœë²„PICK",
        "sortCode": "POPULAR",
        "isUsed": "false",
        "page": page,
        "size": 60,
        "testGroup": "",
        "seen": 120,
        "seenAds": "",
        "caller": "SEARCH"
    }

    print(f"ğŸ“¦ í˜ì´ì§€ {page} ìš”ì²­ ì¤‘...")
    response = requests.get(url, params=params, headers=headers)

    if response.status_code == 200:
        data = response.json()
        items = data.get("data", {}).get("list", [])
        all_items.extend(items)
        print(f"âœ… {len(items)}ê°œ ìƒí’ˆ ìˆ˜ì§‘ ì™„ë£Œ")
    else:
        print(f"âŒ ìš”ì²­ ì‹¤íŒ¨ (í˜ì´ì§€ {page}): {response.status_code}")
    
    # ì„œë²„ ë¶€ë‹´ ì™„í™”
    sleep(1)

# âœ… 5. DataFrame ë³€í™˜
df_musinsa_page = pd.DataFrame(all_items)

# âœ… 6. ì£¼ìš” ì»¬ëŸ¼ ì„ íƒ (í•„ìš” ì‹œ)
df_musinsa_page = df_musinsa_page[[
    "goodsNo", "goodsName", "displayGenderText", "isSoldOut", "normalPrice", 
    "price", "couponPrice", "saleRate", "couponSaleRate", "hasOptionPrice", 
    "brand", "brandName", "brandLinkUrl", "reviewCount", "reviewScore", 
    "isOptionVisible", "isAd", "adTooltip", "plusDeliveryGuideText", 
    "goodsLinkUrl", "thumbnail", "infoLabelList", "imageLabelList", 
    "score", "usedConditionGrade"
]]
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import re

def get_like_count(url):
    """ë¬´ì‹ ì‚¬ ë¸Œëœë“œ í˜ì´ì§€ì—ì„œ ì¢‹ì•„ìš” ìˆ˜ë¥¼ floatìœ¼ë¡œ ë°˜í™˜"""
    try:
        options = Options()
        options.add_argument("--headless")
        driver = webdriver.Chrome(options=options)
        driver.get(url)
        time.sleep(2)
        
        soup = BeautifulSoup(driver.page_source, "html.parser")
        like_elem = soup.select_one("button[class*='like'] span")
        
        if like_elem:
            text = like_elem.text.strip()  # ì˜ˆ: "32.8ë§Œ"
            text = text.replace(",", "")  # ì‰¼í‘œ ì œê±°

            # 'ë§Œ' ë‹¨ìœ„ ì²˜ë¦¬
            if "ë§Œ" in text:
                num = float(re.sub(r"[^0-9.]", "", text))
                return round(num * 10000, 2)
            else:
                return float(re.sub(r"[^0-9.]", "", text))
        else:
            return None
    finally:
        driver.quit()
```
