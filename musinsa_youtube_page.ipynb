## 1. 무신사×유튜버PICK 기획전 페이지 크롤링

- 링크 : https://www.musinsa.com/search/goods?keyword=%EB%AC%B4%EC%8B%A0%EC%82%AC%C3%97%EC%9C%A0%ED%8A%9C%EB%B2%84PICK&gf=M

```python
import requests
# import json
import pandas as pd
from time import sleep

# ✅ 1. 기본 URL
url = "https://api.musinsa.com/api2/dp/v1/plp/goods"

# ✅ 2. 헤더 (브라우저 요청과 동일하게 설정)
headers = {
    "authority": "api.musinsa.com",
    "method": "GET",
    "scheme": "https",
    "accept": "application/json, text/plain, */*",
    "accept-encoding": "gzip, deflate, br, zstd",
    "accept-language": "ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7",
    "origin": "https://www.musinsa.com",
    "referer": "https://www.musinsa.com/",
    "sec-fetch-dest": "empty",
    "sec-fetch-mode": "cors",
    "sec-fetch-site": "same-site",
    "user-agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/118.0.0.0 Safari/537.36"
    ),
}

# ✅ 3. 크롤링 결과를 저장할 리스트
all_items = []

# ✅ 4. 페이지 1~8 반복
for page in range(1, 9):
    params = {
        "gf": "A",
        "keyword": "무신사×유튜버PICK",
        "sortCode": "POPULAR",
        "isUsed": "false",
        "page": page,
        "size": 60,
        "testGroup": "",
        "seen": 120,
        "seenAds": "",
        "caller": "SEARCH"
    }

    print(f"📦 페이지 {page} 요청 중...")
    response = requests.get(url, params=params, headers=headers)

    if response.status_code == 200:
        data = response.json()
        items = data.get("data", {}).get("list", [])
        all_items.extend(items)
        print(f"✅ {len(items)}개 상품 수집 완료")
    else:
        print(f"❌ 요청 실패 (페이지 {page}): {response.status_code}")
    
    # 서버 부담 완화
    sleep(1)

# ✅ 5. DataFrame 변환
df_musinsa_page = pd.DataFrame(all_items)

# ✅ 6. 주요 컬럼 선택 (필요 시)
df_musinsa_page = df_musinsa_page[[
    "goodsNo", "goodsName", "displayGenderText", "isSoldOut", "normalPrice", 
    "price", "couponPrice", "saleRate", "couponSaleRate", "hasOptionPrice", 
    "brand", "brandName", "brandLinkUrl", "reviewCount", "reviewScore", 
    "isOptionVisible", "isAd", "adTooltip", "plusDeliveryGuideText", 
    "goodsLinkUrl", "thumbnail", "infoLabelList", "imageLabelList", 
    "score", "usedConditionGrade"
]]
```

## 2. brand별 좋아요 수를 집계하기 위한 웹 크롤링

```python
# url하나에 대한 test code
# from selenium import webdriver
# from selenium.webdriver.chrome.options import Options
# from bs4 import BeautifulSoup
# import time

# url = "https://www.musinsa.com/brand/corrs"  # 브랜드 URL

# options = Options()
# options.add_argument("--headless")
# driver = webdriver.Chrome(options=options)

# driver.get(url)
# time.sleep(2)  # 렌더링 대기

# soup = BeautifulSoup(driver.page_source, "html.parser")

# # 좋아요 숫자 요소 선택
# like = soup.select_one("button[class*='like'] span")  # ❤️ 옆 숫자 영역
# print("좋아요 수:", like.text.strip())

# driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import re

def get_like_count(url):
    """무신사 브랜드 페이지에서 좋아요 수를 float으로 반환"""
    try:
        options = Options()
        options.add_argument("--headless")
        driver = webdriver.Chrome(options=options)
        driver.get(url)
        time.sleep(2)
        
        soup = BeautifulSoup(driver.page_source, "html.parser")
        like_elem = soup.select_one("button[class*='like'] span")
        
        if like_elem:
            text = like_elem.text.strip()  # 예: "32.8만"
            text = text.replace(",", "")  # 쉼표 제거

            # '만' 단위 처리
            if "만" in text:
                num = float(re.sub(r"[^0-9.]", "", text))
                return round(num * 10000, 2)
            else:
                return float(re.sub(r"[^0-9.]", "", text))
        else:
            return None
    finally:
        driver.quit()
```

- 우리는 유튜브x기획전에 있는 브랜드의 좋아요 수를 집계하는 것이 목표

```python
df_musinsa_ubrand = df_musinsa_page.drop_duplicates(subset=['brandName'], keep='first')
```

```python
df_musinsa_ubrand
```

```python
df_musinsa_ubrand["likeCount"] = df_musinsa_ubrand["brandLinkUrl"].apply(get_like_count)
```
