## musinsa_label과 musinsa_counts Data Join

```python
# musinsa_label_data 불러오기
import pandas as pd

# 1. 파일 경로 지정
file_path_musinsa_label = "./Data/musinsa_label_data.csv"

# 2. CSV 파일을 DataFrame으로 불러오기
df_musinsa_label = pd.read_csv(file_path_musinsa_label, encoding="utf-8")  # 보통 utf-8, 혹은 cp949 필요할 수 있음

# 3. 데이터 확인
print(df_musinsa_label.head())
```

```python
# musinsa_counts_data 불러오기
import pandas as pd

# 1. 파일 경로 지정
file_path_musinsa_counts = "./Data/musinsa_counts_data.csv"

# 2. CSV 파일을 DataFrame으로 불러오기
df_musinsa_counts = pd.read_csv(file_path_musinsa_counts, encoding="utf-8")  # 보통 utf-8, 혹은 cp949 필요할 수 있음

# 3. 데이터 확인
print(df_musinsa_counts.head())
```

```python
# 두 DataFrame join
df_musinsa = pd.merge(
    df_musinsa_label,      # 왼쪽 DataFrame
    df_musinsa_counts,             # 오른쪽 DataFrame
    left_on='goodsNo',     # df_musinsa_label 기준 컬럼
    right_on='relationId', # df_musinsa_counts 기준 컬럼
    how='inner'            # inner join (양쪽 모두 있는 데이터만)
)

# 결과 확인
print(df_musinsa.shape)
print(df_musinsa.head())
```
