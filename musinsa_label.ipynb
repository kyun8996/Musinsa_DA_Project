## 1. drawfit 상품 정보 크롤링하기 (GET방식)

```python
import requests
import time
import pandas as pd
import random

all_datas_drawfit = []
page_num = 1

url = "https://api.musinsa.com/api2/dp/v1/plp/goods"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Referer": "https://www.musinsa.com/",
    "Accept": "application/json",
}

while True:
    params_drawfit = {
        "gf": "A",
        "sortCode": "SALE_ONE_MONTH_AMOUNT",
        "brand": "drawfit",             # 브랜드 명에 따라 변경
        "page": page_num,
        "size": 30,
        "caller": "FLAGSHIP",
    }

    res_drawfit = requests.get(url, headers=headers, params=params_drawfit)
    print(f"Page {page_num} → Status: {res_drawfit.status_code}")

    if res_drawfit.status_code != 200:
        print("요청실패 - 수집중단")
        break

    data_drawfit = res_drawfit.json()
    items = data_drawfit.get("data", {}).get("list", [])
    
    if not items:
        print("마지막 페이지 도달")
        break

    all_datas_drawfit.extend(items)
    page_num += 1

    random_delay = random.randint(1, 10) 
    time.sleep(random_delay)
    
  
# 데이터프레임에 저장
df_drawfit = pd.DataFrame(all_datas_drawfit)
print(df_drawfit.head())
print(f"\n총 상품 수: {len(df_drawfit)}개")
```
